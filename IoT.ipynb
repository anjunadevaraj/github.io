```python
import matplotlib.pyplot as plt

# Create a figure
fig, ax = plt.subplots(figsize=(8, 6))

# Define elements of the IoT module
elements = {
    "CMP213 - Internet of Things": (0.5, 0.9),
    "Sensors & Actuators": (0.2, 0.7),
    "Embedded Devices": (0.8, 0.7),
    "IoT System Constraints & Security": (0.2, 0.5),
    "Cloud Computing & IoT": (0.8, 0.5),
    "IoT Programming Frameworks": (0.5, 0.4),
    "Design & Implementation": (0.5, 0.3),
    "Testing & Evaluation": (0.5, 0.2),
    "Final Project - IoT Smart Solution": (0.5, 0.1)
}

# Plot module components as nodes
for label, (x, y) in elements.items():
    ax.text(x, y, label, ha='center', va='center', fontsize=9,
            bbox=dict(facecolor='lightblue', edgecolor='black', boxstyle='round,pad=0.3'))

# Draw connections between components
connections = [
    ("CMP213 - Internet of Things", "Sensors & Actuators"),
    ("CMP213 - Internet of Things", "Embedded Devices"),
    ("Sensors & Actuators", "IoT System Constraints & Security"),
    ("Embedded Devices", "Cloud Computing & IoT"),
    ("IoT System Constraints & Security", "IoT Programming Frameworks"),
    ("Cloud Computing & IoT", "IoT Programming Frameworks"),
    ("IoT Programming Frameworks", "Design & Implementation"),
    ("Design & Implementation", "Testing & Evaluation"),
    ("Testing & Evaluation", "Final Project - IoT Smart Solution")
]

# Draw arrows between components
for start, end in connections:
    start_x, start_y = elements[start]
    end_x, end_y = elements[end]
    ax.arrow(start_x, start_y - 0.03, end_x - start_x, end_y - start_y + 0.03,
             head_width=0.02, head_length=0.02, fc='gray', ec='gray', alpha=0.6)

# Remove axes
ax.set_xticks([])
ax.set_yticks([])
ax.set_frame_on(False)

# Set title
ax.set_title("CMP213 - Internet of Things Module Overview", fontsize=12, fontweight='bold')

# Show diagram
plt.show()
```
